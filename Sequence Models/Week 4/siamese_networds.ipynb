# Siamese Newtork

This is a neural network, which is made up of two identical neural networks, that are merged in the end. They can be used to figure out :
- Question Duplicates
Siamese Networks, can be used to check for question duplicates.
- Authenticate Handwritten Checks
- Query Duplicates

## Architecture of Siamese Network

They have two identical sub-networks(sister networks, which merge together to produce a similarity score). If the similairity is less than some threshold, then we say that the questinos are different. This threshold is the count of how often you would want to test for similarity.



## Cost Function for a Siamese Network

Sentences are compared to an anchor sentence, and if a sentence is similar it is positive w.r.t. the anchor, and if it is negative, it is negative w.r.t the anchor.

diff = s(A,N) - s(A-P)

## Triplets

How old are you? Anchor

What is your age? Positive

Where are you from? Negative

Simple Loss : diff = s(A,N) - s(A,P)

For efficient training, we select the problematic triplets, or hard triplets to better train and focus the model on areas it is performing poorly.

## Computing the Cost Function. Hard Negative Mining

1. Prepare the data in batches, which have duplicates as pairs. The number of pairs is known as the batch size. No two sentences in the column are duplicates.
2. Prepare a matrix. The diagonals represnet the similarities for the duplicate pairs.

- Mean negative : mean of off-diagonal values in each row
- Closes negative : off diagonal value closes to (but less than) the value on the diagonal in each row.

Finally, define a full loss, which is the sum of both of these losees.
